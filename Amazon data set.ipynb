```python
import mysql.connector
import pandas as pd
from sqlalchemy import create_engine
import re

# Database configuration
db_config = {
    'host': 'localhost',
    'user': 'root',
    'password': '@11Antasystic',
    'database': 'ecommerce_db'
}

import pandas as pd

def clean_amazon_data(filepath="C:\\Users\\USER\\OneDrive\\Documents\\Wanjiru\\amazon.csv"):
    """Cleans the Amazon dataset."""
    try:
        df = pd.read_csv(filepath)  # Corrected indentation and using filepath
        return df #added return statement.
    except FileNotFoundError:
        print(f"Error: File not found at {filepath}")
        return None

    price_cols = ['discounted_price', 'actual_price']
    for col in price_cols:
        if col in df.columns:
            df[col] = df[col].astype(str).str.replace(r'[^\d.]', '', regex=True)
            df[col] = pd.to_numeric(df[col], errors='coerce')

    if 'rating_count' in df.columns:
        df['rating_count'] = df['rating_count'].astype(str).str.replace(',', '', regex=False)
        df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce').astype('Int64')

    if 'category' in df.columns:
        df['category'] = df['category'].astype(str).str.split('|').str[0]

    if 'actual_price' in df.columns and 'discounted_price' in df.columns:
        df['discount_amount'] = df['actual_price'] - df['discounted_price']

    if 'discount_percentage' in df.columns:
        df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%', '', regex=False)
        df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')

    for col in df.columns:
        if 'review' in col.lower() and ('date' in col.lower() or 'time' in col.lower()):
            try:
                df['review_year'] = pd.to_datetime(df[col], errors='coerce').dt.year
                break
            except (TypeError, ValueError):
                pass

    return df

def load_data_to_mysql(df, user, password, host, database):
    """Loads cleaned DataFrame into a MySQL table."""
    try:
        encoded_password = urllib.parse.quote_plus(password)
        engine = create_engine(f"mysql+mysqlconnector://{user}:{encoded_password}@{host}/{database}")
        df.to_sql(name='products', con=engine, if_exists='replace', index=False)
        print("Data loaded successfully.")
    except Exception as e:
        print(f"Error loading data: {e}")

# Example of using the function:
data = {'col1': [1, 2], 'col2': [3, 4]}
df = pd.DataFrame(data)

load_data_to_mysql(df, db_config['user'], db_config['password'], db_config['host'], db_config['database'])

# 4. Create a MySQL Database (and load data)
cleaned_df = clean_amazon_data("C:\\Users\\USER\\OneDrive\\Documents\\Wanjiru\\amazon.csv") 
if cleaned_df is not None:
    load_data_to_mysql(cleaned_df, db_config['user'], db_config['password'], db_config['host'], db_config['database'])
```

```python
import mysql.connector
import pandas as pd

# Database connection parameters
config = {
    'host': 'localhost',
    'user': 'root',
    'password': '@11Antasystic',
    'database': 'ecommerce_db'
}

# SQL query to retrieve top 10 most reviewed products
query_top_10_most_reviewed = """
SELECT 
    product_name, 
    rating_count 
FROM 
    products 
ORDER BY 
    rating_count DESC 
LIMIT 10;
"""

try:
    # Establish MySQL connection
    db = mysql.connector.connect(**config)
    cursor = db.cursor()

    # Execute the query
    cursor.execute(query_top_10_most_reviewed)

    # Fetch results into a DataFrame
    df_top_10_most_reviewed = pd.DataFrame(
        cursor.fetchall(), 
        columns=['product_name', 'rating_count']
    )

    # Display the DataFrame
    print("Top 10 Most Reviewed Products:")
    print(df_top_10_most_reviewed)

except mysql.connector.Error as e:
    print(f"MySQL Error: {e}")

finally:
    # Close cursor and connection
    if 'cursor' in locals() and cursor is not None:
        cursor.close()
    if 'db' in locals() and db.is_connected():
        db.close()
        print("\nMySQL connection closed.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("C:\\Users\\USER\\OneDrive\\Documents\\Wanjiru\\amazon.csv")

df['discount_percentage'] = pd.to_numeric(df['discount_percentage'].str.replace('%', ''), errors='coerce')
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df_cleaned = df.dropna(subset=['discount_percentage', 'rating'])

plt.figure(figsize=(10, 6))
sns.regplot(
    x='discount_percentage', 
    y='rating', 
    data=df_cleaned, 
    scatter_kws={'alpha': 0.3, 'color': 'blue'}, 
    line_kws={'color': 'red', 'linestyle': '--'}
)

correlation = df_cleaned[['discount_percentage', 'rating']].corr().iloc[0, 1]
plt.title(f'Discount Percentage vs. Rating (Correlation: {correlation:.2f})')
plt.xlabel('Discount Percentage (%)')
plt.ylabel('Rating (1-5 stars)')
plt.grid(linestyle='--', alpha=0.7)
plt.show()
```
