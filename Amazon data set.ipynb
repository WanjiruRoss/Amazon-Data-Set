```python
import mysql.connector as msql
from mysql.connector import Error
import pandas as pd

# Load your DataFrame 
df_cleaned = pd.read_csv("C:\\Users\\USER\\OneDrive\\Documents\\Wanjiru\\amazon.csv")

try:
    # Connect to MySQL
    conn = msql.connect(
        host='localhost',
        user='root',
        password='@11Antasystic',
        database='ecommerce_db'
    )

    if conn.is_connected():
        cursor = conn.cursor()
        cursor.execute("SELECT database();")
        record = cursor.fetchone()
        print("You're connected to database:", record)

        # Drop table if exists
        cursor.execute('DROP TABLE IF EXISTS products;')  # Fixed table name
        print('Table dropped (if existed).')

        # Create table
        print('Creating table...')
        cursor.execute("""
            CREATE TABLE IF NOT EXISTS products (
                product_id VARCHAR(255) PRIMARY KEY,
                product_name TEXT,
                category TEXT,
                discounted_price FLOAT,
                actual_price FLOAT,
                discount_percentage FLOAT,
                rating FLOAT,
                rating_count INT,
                about_product TEXT,
                user_id TEXT,
                user_name TEXT,
                review_id TEXT,
                review_title TEXT,
                review_content TEXT,
                img_link TEXT,
                product_link TEXT,
                main_category TEXT,
                sub_category TEXT,
                discount_amount FLOAT
            )
        """)  # Fixed syntax
        print("Table created successfully.")

        # Prepare data for insertion
        df_cleaned = df_cleaned.where(pd.notnull(df_cleaned), None)  # Convert NaN to None
        data_tuples = [tuple(x) for x in df_cleaned.to_records(index=False)]

        # Insert data using execute many (more efficient)
        sql = """
            INSERT INTO products VALUES (
                %s,%s,%s,%s,%s,%s,%s,%s,%s,%s,
                %s,%s,%s,%s,%s,%s,%s,%s,%s
            )
        """
        cursor.executemany(sql, data_tuples)
        conn.commit()
        print(f"Successfully inserted {len(df_cleaned)} records")

except Error as e:
    print("Error while connecting to MySQL:", e)
    if conn.is_connected():
        conn.rollback()

finally:
    if 'cursor' in locals() and cursor is not None:
        cursor.close()
    if conn.is_connected():
        conn.close()
        print("MySQL connection closed.")
```

```python
import mysql.connector
import pandas as pd

# Database connection parameters
config = {
    'host': 'localhost',
    'user': 'root',
    'password': '@11Antasystic',
    'database': 'ecommerce_db'
}

# SQL query to retrieve top 10 most reviewed products
query_top_10_most_reviewed = """
SELECT 
    product_name, 
    rating_count 
FROM 
    products 
ORDER BY 
    rating_count DESC 
LIMIT 10;
"""

try:
    # Establish MySQL connection
    db = mysql.connector.connect(**config)
    cursor = db.cursor()

    # Execute the query
    cursor.execute(query_top_10_most_reviewed)

    # Fetch results into a DataFrame
    df_top_10_most_reviewed = pd.DataFrame(
        cursor.fetchall(), 
        columns=['product_name', 'rating_count']
    )

    # Display the DataFrame
    print("Top 10 Most Reviewed Products:")
    print(df_top_10_most_reviewed)

except mysql.connector.Error as e:
    print(f"MySQL Error: {e}")

finally:
    # Close cursor and connection
    if 'cursor' in locals() and cursor is not None:
        cursor.close()
    if 'db' in locals() and db.is_connected():
        db.close()
        print("\nMySQL connection closed.")
```
